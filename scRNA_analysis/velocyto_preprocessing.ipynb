# RNA velocity preprocessing (sex-chimeric dataset)

This notebook prepares Velocyto loom outputs for RNA velocity analysis, including:

- **Loading Velocyto loom files** for all samples.
- **Making barcode and gene names unique**, then **concatenating** samples with batch annotation.
- **Filtering to the macrophage subset** and **transferring metadata** from the scRNA analysis.
- **Merging velocity matrices** into the main AnnData object for downstream velocity estimation (Part 2).

### Rename cell barcodes

Make barcode names unique across samples so the concatenated object has no index collisions.

```python
for i, ldata1 in enumerate([WR1_WT_WT_sample1, WR1_WT_WT_sample2, WR1_WT_WT_sample3, WR1_WT_WT_sample4, WR1_mdx_WT_sample1, WR1_mdx_WT_sample2, WR1_mdx_WT_sample3, WR1_mdx_WT_sample4, WR1_mdx_WT_sample5,
                            WR2_WT_WT_sample1, WR2_WT_WT_sample2, WR2_WT_WT_sample3, WR2_WT_WT_sample4, WR2_mdx_WT_sample1, WR2_mdx_WT_sample2, WR2_mdx_WT_sample3, WR2_mdx_WT_sample4, WR2_mdx_WT_sample5,
                            MR1_WT_mdx_sample1, MR1_mdx_mdx_sample1, MR1_mdx_mdx_sample2, MR2_WT_mdx_sample1, MR2_WT_mdx_sample2, MR2_mdx_mdx_sample1]):
    barcodes = [bc.split(':')[1] for bc in ldata1.obs.index.tolist()]
    barcodes = [bc[0:len(bc)-1] + f'-1-{i}' for bc in barcodes]
    ldata1.obs.index = barcodes
```

## Transfer metadata

Transfer cluster, sex, and condition annotations from the integrated object.

```python
results_file = './sex_chimeric_only_Macro.h5ad'
adata = sc.read(results_file)
cells_to_keep = adata.obs.index
adata_merge_filtered = adata_merge[cells_to_keep, :].copy()
```

```python
for col in ['WT_sum_cat', 'pred_sex', 'leiden_Macro']:
    if col in adata.obs.columns:  # ensure column exists
        adata_merge_filtered.obs[col] = adata.obs.loc[adata_merge_filtered.obs.index, col]

if 'leiden_Macro_colors' in adata.uns:
    adata_merge_filtered.uns['leiden_Macro_colors'] = adata.uns['leiden_Macro_colors']

adata_merge_filtered.write('sex_chimeric_raw_scVelo.h5ad')
```

```python
results_file = './sex_chimeric_raw_scVelo.h5ad'
adata = sc.read(results_file)
def clean_index(index):
    parts = index.split('-')
    new_parts = []
    for i, part in enumerate(parts):
        if i == 0 or part != parts[i-1]:
            new_parts.append(part)
    return '-'.join(new_parts)

new_index = ldata.obs.index.map(clean_index)
ldata.obs.index = new_index
```

## Merge velocity matrices into AnnData

Merge the concatenated loom (velocity) matrices into the main AnnData object and attach the UMAP embedding for visualization.

```python
# merge matrices into the original adata object
adata_merged = scv.utils.merge(adata, ldata)
results_file = './sex_chimeric_celldancer_corr.h5ad'
adata_b4 = sc.read(results_file)

adata_merged.obsm['X_umap'] = adata_b4.obsm['X_umap']
```
